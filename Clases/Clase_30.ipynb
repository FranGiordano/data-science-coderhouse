```python
def promedio_pais(fila):
  suma = cantidad = 0
  for col in fila:
    if str(col)[-1] == "k":
      col = col[:-1]
    suma += float(col)
    cantidad += 1
  return suma/cantidad

import numpy as np
df_pais=df_pais.apply(lambda x: x.replace("k", ""))
df_pais
```
